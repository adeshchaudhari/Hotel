```python
col_name = ["avg_price_per_room", "lead_time"]
new = df2[col_name]
std = StandardScaler()
arr1 = std.fit_transform(new)
df3 = pd.DataFrame(arr1, columns= new.columns)
sns.distplot(df3["avg_price_per_room"])
```

```python
x = df2.drop("booking_status", axis=1)
y = df2["booking_status"]
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=20, stratify=y)
```

```python
x_train_sca = std.fit_transform(x_train)
x_test_sca = std.transform(x_test)
```

```python
np.sqrt(df2.shape[0])
```

```python
knn = KNeighborsClassifier(n_neighbors=189)
knn.fit(x_train_sca, y_train)
```

```python
def evaluation(string,model, ind_var,y_act):
    print(string.center(50, '*'))
    
    pred = model.predict(ind_var)
    
    cnf_matrix = confusion_matrix(y_act, pred)
    print('Confusion Matrix :\n', cnf_matrix)
    
    print('**'*25)
    acc_score = accuracy_score(y_act, pred)
    print('Accuracy is :', acc_score*100)
    print('**'*25)
    
    clf_report = classification_report(y_act, pred)
    print('Classification Report :\n', clf_report)
    
evaluation('Train Data Evaluation', knn, x_train_sca, y_train)
```

```python
evaluation('Test Data Evaluation', knn, x_test_sca, y_test)
```

```python
with open ("Hotel_price_KNN.pkl", "wb")as f:
    pickle.dump(knn,f)
```
